```python
import os
import glob
import pandas as pd
```

```python
# indicate file path and look for csv
os.chdir("../data")
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
# combine 12 csv files into one
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
# read combined csv into a datframe

trip_data = "../data/combined_csv.csv"
trip_data_df = pd.read_csv(trip_data, encoding="ISO-8859-1")
trip_data_df.head()
```

```python
# check for nulls
trip_data_df.count()
```

```python
# select only needed columns
# I would like to keep all the columns, but this makes the dataset smaller and more managable.

trip_data_df = trip_data_df[['ï»¿tripduration','starttime','start station id', 'start station name','start station latitude','start station longitude','usertype','bikeid','birth year','gender']]
trip_data_df.head()
```

```python
# rename columns

trip_data_df.rename(columns = {'ï»¿tripduration':'duration(s)', 'starttime':'start_time', 'start station id': 'start_id',
                             'start station name': 'start_name', 'start station latitude': 'start_lat', 'start station longitude': 'start_lng',
                             'bikeid': 'bike_id', 'birth year': 'birth_year'}, inplace = True)
trip_data_df.head()
```

```python
# rearrange columns
trip_data_df = trip_data_df[['bike_id','usertype','birth_year','gender', 'start_lat', 'start_lng', 'start_id','start_name', 'duration(s)','start_time']]
trip_data_df.head()
```

```python
# split date from time in 'start_time'
trip_data_df[['date','time']] = trip_data_df.start_time.str.split(expand=True)
trip_data_df.head()
```

```python
# drop start_time and time columns because I only need the date
trip_data_df.drop(trip_data_df.columns[[9, 11]], axis = 1, inplace = True) 
trip_data_df.head()
```

```python
# export as csv for use in Tableau

trip_data_df.to_csv("../clean_data/2020.csv", index=False, header=True)
```
